### Escalado de características


Estos datos contienen valores númericos con magnitudes grandes entre las columnas numéricas, como: `f0`, `f1` y `f2`. Lo cual, el algoritmo puede encontrar que alguna sea más importante que otra. Por consiguiente, se realizará un escalado de carácterísticas.

#### Region 1

```python
#Se establecen las columnas numericas que sera escaladas para OHE
numeric = ['f0', 'f1', 'f2']

scaler = StandardScaler()
scaler.fit(features_train_1[numeric])
```

```python
features_train_1[numeric]= scaler.transform(features_train_1[numeric])
features_valid_1[numeric] = scaler.transform(features_valid_1[numeric])
```

#### Region 2

```python
#Se establecen las columnas numericas que sera escaladas para OHE
numeric = ['f0', 'f1', 'f2']

scaler = StandardScaler()
scaler.fit(features_train_2[numeric])
```

```python
features_train_2[numeric]= scaler.transform(features_train_2[numeric])
features_valid_2[numeric] = scaler.transform(features_valid_2[numeric])
```

#### Region 3

```python
#Se establecen las columnas numericas que sera escaladas para OHE
numeric = ['f0', 'f1', 'f2']

scaler = StandardScaler()
scaler.fit(features_train_3[numeric])
```

```python
#Escalado para OHE
features_train_3[numeric]= scaler.transform(features_train_3[numeric])
features_valid_3[numeric] = scaler.transform(features_valid_3[numeric])
```

## Creación del modelo

Se creará un modelo de regresión lineal

### Region 1

```python
model = LinearRegression()  
model.fit(features_train_1, target_train_1) 
predicted_valid_1 = model.predict(features_valid_1) 

recm_valid_1 = mean_squared_error(target_valid_1, predicted_valid_1) ** 0.5 #Se eleva a potencia 0.5 para obtener RECM
r2_valid_1 = r2_score(target_valid_1, predicted_valid_1)
product_mean_1 = target_valid_1.mean()

print("Volumnen promedio de las reservas previstas", product_mean_1)
print("Raiz cuadrada del error cuadrático medio del conjunto de validación:", recm_valid_1)
print("Valor R2 del conjuntos de validación:", r2_valid_1)
```

Se puede observar que modelo no predice de manera adecuada, se tiene un valor raiz cuadrada de error cuadrático medio de 37,000 barriles, esto quiere decir que que en promedio las predicciones tienen un error medio de más/menos el RECM. 

Este cálculo se valida con el R2 del 27%, querie decir que el modelo predice, pero de manera poco óptima.

### Region 2

```python
model.fit(features_train_2, target_train_2) 
predicted_valid_2 = model.predict(features_valid_2) 

recm_valid_2 = mean_squared_error(target_valid_2, predicted_valid_2) ** 0.5 #Se eleva a potencia 0.5 para obtener RECM
r2_valid_2 = r2_score(target_valid_2, predicted_valid_2)
product_mean_2 = target_valid_2.mean()

print("Volumnen promedio de las reservas previstas", product_mean_2)
print("Raiz cuadrada del error cuadrático medio del conjunto de validación:", recm_valid_2)
print("Valor R2 del conjuntos de validación:", r2_valid_2)
```

Se puede notar que el modelo genera buenas predicciones, debido a que se tiene una raiz cuadrada del error cuadrático medio de menos de 1,000 barriles.

El valor R2 del 99, lo que quiere decir que las predicciones del modelo son acertadas y tiene muy poco error con respecto a los conjuntos de validación y test.

### Region 3

```python
model.fit(features_train_3, target_train_3) 
predicted_valid_3 = model.predict(features_valid_3) 

recm_valid_3 = mean_squared_error(target_valid_3, predicted_valid_3) ** 0.5 #Se eleva a potencia 0.5 para obtener RECM
r2_valid_3 = r2_score(target_valid_3, predicted_valid_3)
product_mean_3 = target_valid_3.mean()

print("Volumnen promedio de las reservas previstas", product_mean_3)
print("Raiz cuadrada del error cuadrático medio del conjunto de validación:", recm_valid_3)
print("Valor R2 del conjuntos de validación:", r2_valid_3)
```

Se puede notar que el modelo no genera buenas predicciones, debido a que se tiene una raiz cuadrada del error cuadrático medio de 40,000 barriles.

El valor R2 del 19%, lo que quiere decir que las predicciones del modelo pueden ser no tan acertadas con un rango de error alto los conjuntos de validación y test.

## Bootstrapping

### Región 1

```python
#Se convierten las predicciones en un series para utilizar la funcion samle
region_1 = pd.Series(predicted_valid_1)

state = np.random.RandomState(931106)

values = []

for i in range(1000):
    subsample = target_valid_1.sample(n=500, replace=True, random_state=state)
    sub_predicted = region_1[subsample.index]
    values.append(total_price(sub_predicted, subsample))

values = pd.Series(values)

lower = values.quantile(0.025)

mean = values.mean()

upper = values.quantile(0.975)

risk = (values < 0).sum() / len(values)*100

print("El límite inferior inferior del intervalo de ganancias:", lower/1000000)
print("El límite superior del intervalo de ganancias:", upper/1000000)
print("El promedio de las ganancias:", mean/1000000)
print("Porcentaje de riesgo:", risk)
```
